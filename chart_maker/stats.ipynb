```python
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from IPython.display import display # to display images in jupyter notebook
import numpy as np
```

```python
# read csv data into dataframe object
df = pd.read_csv("pioneer_logger.csv")
df.head()
```

```python
# replace collision's null values w/ 0
df.replace({'field.collision.x':-100000, 'field.collision.y':-100000},0,inplace=True)
```

```python
# remove unnecessary cols
df=df.drop(['field.header.frame_id'],axis=1)
```

```python
df.describe()
```

```python
df_pos = df[['field.robot_pos.x','field.robot_pos.y']]
df_pos = df_pos.rename(columns={'field.robot_pos.x':'robot_pos.x','field.robot_pos.y':'robot_pos.y'})
```

```python
# plot robot's path
df_pos.plot.scatter(x='robot_pos.x',y='robot_pos.y')
```

```python
df_pos.plot.scatter(x='robot_pos.x',y='robot_pos.y',s=0.5)
```

```python
path = df_pos.plot.scatter(x='robot_pos.x',
                           y='robot_pos.y',
                           xlim=(-1,8),
                           ylim=(-2.0,3.00),
                           figsize=(9,7),
                           s=10,
                           c='cornflowerblue')
path.get_figure().savefig("robot_pos2.png")
```

```python
# remove axis and save for superimposing on map
path.axis('off')
display(path.get_figure())
path.get_figure().savefig("robot_pos2_no_axis.png", transparent=True)
```

```python
# load map and path_no_axis images for merging
# map is background, path_no_axis is foreground
background = Image.open("level1.pgm")
foreground = Image.open("robot_pos2_no_axis.png")
```

```python
# view map
display(background)
background.size
```

```python
help(Image)
```

```python
xp = 0.253819
yp = 0.008350
bx , by = background.size
```

```python
#crop background map
# (left, upper, right, lower) = (bx, 0, 300, 230)
# background_c = background.crop((left, upper, right, lower))
# display(background_c)
for idx, row in df_pos.iterrows():
    print(row['robot_pos.x'], row['robot_pos.y'])
```

```python
#background.getpixel((20,20))
resolution = 0.100000
xn = int((xp/resolution) + bx/2)
yn = int((yp/resolution) + by/2)
print(xn,yn)
for i in range(20):
    background.putpixel((xn+i,yn+i),2)
display(background)
```

```python
# data from level1.yaml
# resolution: 0.100000
# origin: [-20.000000, -20.000000, 0.000000]
# ref: https://wiki.nps.edu/pages/viewpage.action?pageId=1157791916
resolution = 0.100000
origin = [-20.000000, -20.000000, 0.000000]
#span of image in meters
dx = background.size[0]*resolution
dy = background.size[1]*resolution
print("dx:{0},dy:{1}".format(dx,dy))

#set size scaling
xWorldLimits = [0 dx] + origin[0]
yWorldLimits = [0 dy] + origin[1]
```

```python
#crop background map
(left, upper, right, lower) = (175, 120, 300, 230)
background_c = background.crop((left, upper, right, lower))
display(background_c)
```

```python
#resize background map
background_cr = background_c.resize((int(foreground.width*1.2),int(foreground.height*1.2)))
display(background_cr)
background_cr.size
```

```python
display(foreground)
foreground.size
```

```python
# paste path_no_axis on map
background_cr.paste(foreground, (60,80), foreground.convert('RGBA'))
display(background_cr)
```

```python
#save the final image
background_cr.save('robot_path_map_super.png')
```
